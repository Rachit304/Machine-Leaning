**5. Apply KNN for the preprocessed dataset use 5 different K Values and 3 different distance measures**

```python
mc.rename(columns = {'3D_available':'Three_D'}, inplace = True)
```

```python
mc.Three_D.replace(('YES', 'NO'), (1, 0), inplace=True)
```

```python
mc.head()
```

```python
X = mc[['Avg_age_actors','Num_multiplex','Start_Tech_Oscar']]
y = mc.Three_D
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, random_state =10,test_size=0.2)
len(X_train)
```

```python
len(X_test)
```

Split the data in 80%

```python
from sklearn.neighbors import KNeighborsClassifier  
knn =KNeighborsClassifier(n_neighbors  =  3)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

**6.Display the confusion matrix, accuracy measures, classification report.**

```python
from sklearn.metrics import confusion_matrix

pred = knn.predict(X_test)

cm = confusion_matrix(y_test,pred)
cm
```

```python
len(y_test)
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm , annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred))
```
